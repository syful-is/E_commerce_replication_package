```python
import re
import pandas as pd
import numpy as np
from collections import Counter
import statistics
import os

os.chdir("F:/1_NAIST_Research_SE/SE_meeting/Ecommerce/")

df1=pd.read_csv("RQ1_ecommerce_version_1.csv",low_memory=False)

Dominant_Topic=[]
Topic_Perc_Contrib=[]
Keywords=[]
Text=[]

for i in range(0, len(df1)):
    if df1['Topic_Perc_Contrib'][i]>=0.50:
        Dominant_Topic.append(df1['Dominant_Topic'][i])
        Topic_Perc_Contrib.append(df1['Topic_Perc_Contrib'][i])
        Keywords.append(df1['Keywords'][i])
        Text.append(df1['Text'][i])
        
dict={'Dominant_Topic':Dominant_Topic, 'Topic_Perc_Contrib':Topic_Perc_Contrib, 'Keywords':Keywords, 'Text':Text}
df2=pd.DataFrame(dict)
df2.to_csv("RQ1_ecommece_TOpic_threshold_050.csv")
```
